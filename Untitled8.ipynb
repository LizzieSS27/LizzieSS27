# Código

Este codigo calcula la matriz de correlación de Kendall.

• La correlación de Kendal evalúa la ordenación de los datos y es útil con conjuntos de datos pequeños o cuando se quiere concordancia entre clasificaciones.

```python
d=['PUNT_LECTURA_CRITICA','PUNT_MATEMATICAS','PUNT_C_NATURALES','PUNT_SOCIALES_CIUDADANAS','PUNT_INGLES']
x1=tabla_regrecion[['PUNT_LECTURA_CRITICA','PUNT_MATEMATICAS','PUNT_C_NATURALES','PUNT_SOCIALES_CIUDADANAS','PUNT_INGLES']]
y=tabla_regrecion['ESTU_INSE_INDIVIDUAL']
i=0
while i< len(d):
  x=x1.drop(d[i],axis=1)
  trainx, testx, trainy, testy = train_test_split(x, y, test_size = 0.2, random_state = 0)
  modelo_lineal=LinearRegression()
  regresion=modelo_lineal.fit(trainx,trainy)
  pred_lin=regresion.predict(testx)
  r2_lin=regresion.score(testx,testy)
  print(r2_lin)
  i=i+1

```

# Código

Este codigo realiza un análisis de regresión lineal multiple para predecir la variable ESTU_INSE-INDIVIDUAL basándose en las otras variables.

• Bucle de iteración: El código itera sobre cada variable predictora, relaizando un análisis de regresión para cada una.

# Valores

•  0.2934 (PUNT_MATEMATICAS)

• 0.2963 (PUNT_C_NATURALES)

• 0.2964 (PUNT_SOCIALES_CIUDADANAS)

• 0.1556 (PUNT_INGLES)

# Análisis

1. La puntuaciones de Ciencias Naturales y de ciencias sociales son los mejores predictores individuales con respecto a Estu_Inse_Estudiantes, con un margen pequeño.

2. La puntuación en ingles tiene un poder predictivo notablemente menor, lo que implica que esta menos relacionado con el contexto socioeconómico a diferencia de las otras variables.

3. Los valores estan alrededor del 30% de la varianza en el estado socioeconómico individual del estudiante. Esto sugiere que existen otros factores importantes que no se toman en cuenta.

4. El análisis se basa en regresiones simples, por lo cual no se consideran interacciones entre variables no lineales que puedan existir.



```python
t=tabla_regrecion[['PUNT_LECTURA_CRITICA','PUNT_MATEMATICAS','PUNT_C_NATURALES','PUNT_SOCIALES_CIUDADANAS','PUNT_INGLES','ESTU_INSE_INDIVIDUAL']]
x=t.drop('ESTU_INSE_INDIVIDUAL',axis=1)
y=tabla_regrecion[['ESTU_INSE_INDIVIDUAL']]
trainx, testx, trainy, testy = train_test_split(x, y, test_size = 0.2, random_state = 0)
modelo_lineal=LinearRegression()
regresion=modelo_lineal.fit(trainx,trainy)
pred_lin=regresion.predict(testx)
r2_lin=regresion.score(testx,testy)
print(r2_lin)
```
